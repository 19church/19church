```python
#30 use function tier(age)

def tier(age) :
  if age < 25 :
    return "Rookies"
  elif age < 32 :
    return "Prime"
  else :
    return "Almost Retired"

import pandas as pd

df = pd.read_csv('soccerplayerhw.csv')

player_age = df[["short_name", "age"]]

l = list()

for i in range(0, player_age.shape[0]) :
  l.append(tier(player_age.iat[
            i,
            player_age.columns.get_loc("age")
          ]
        )
      )

player_age["tier"] = l

player_age
```
